# **spotify Dataset 1921-2020, 160k+ Tracks **




File data ini berisi data track yang dikelompokkan berdasarkan tahun rilis setiap track, dan memungkinkan pengoperasian deret waktu dilakukan. Setiap baris mewakili satu tahun, setiap kolom mewakili fitur audio. Bias data seminimal mungkin.

## Pertanyaan Analisis



*   1. pada tahun berapa pajang trek tertinggi?
*   2. Perbandingan Panjang trek dalam milidetik (ms) dengan Panjang relatif trek yang berisi segala jenis suara manusia



## Data Preparation

```
import pandas as pd
import numpy as np

spotify = pd.read_csv("/content/data_by_year.csv")
print(spotify.keys())
spotify
```

```
# Pada tahun berapa pannjang terk tertinggi
spotify1=pd.DataFrame(spotify[['year','duration_ms']])
spotify1.sort_values(by=["duration_ms"])
```

```
# Perbandingan Panjang trek dalam milidetik (ms) dengan Panjang relatif trek yang berisi segala jenis suara manusia
spotify2=pd.DataFrame(spotify[['year','duration_ms','speechiness']])
spotify2.sort_values(by=["year"])
```

## Data Visualization

```
import plotly.express as px
import plotly.graph_objects as go
```

```

# 1. Pada tahun berapa panjang trek tertinggi
gbr_spotily1 = px.bar(spotify1,
               x='year',
               y='duration_ms')

gbr_spotily1.update_layout(
    title='Diagram Batang Panjang trek sepanjang tahun 1921 ',
    title_font_color='blue',
    title_font_size=30,
    xaxis=dict(
        title='year',
        titlefont_size=16,
        tickfont_size=14),
    yaxis=dict(
        title='duration_ms',
        titlefont_size=16,
        tickfont_size=14))

gbr_spotily1.show()
```
